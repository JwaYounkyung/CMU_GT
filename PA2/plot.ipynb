```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open('results/maddpg_simple/test_agrewards2.pkl', 'rb') as f:
    data = pickle.load(f)

print(data)

def result_plot(data):
    plt.clf() 
    plt.figure(figsize=(15, 5))

    epochs = range(1000, len(data)*1000 + 1,1000)

    plt.plot(epochs, data, label='Average Reward', marker='o')
    plt.title('Average Reward on the Simple Scenario')
    plt.legend()

    plt.show()

result_plot(data)
```

```python
with open('results/maddpg_uav/test_agrewards2.pkl', 'rb') as f:
    data = pickle.load(f)

agent1, agent2, agent3 = [], [], []
for i in range(0, int(len(data)/3)):
    agent1.append(data[3*i])
    agent2.append(data[3*i+1])
    agent3.append(data[3*i+2])
print(agent1, agent2, agent3)
print(data)

def result_plot(data):
    plt.clf() 
    plt.figure(figsize=(15, 5))

    epochs = range(500, len(agent1)*500 + 1,500)

    plt.plot(epochs, agent1, label='poacher', marker='o')
    plt.plot(epochs, agent2, label='ranger', marker='o')
    plt.plot(epochs, agent3, label='UAV', marker='o')
    plt.title('Average Reward on the Ranger-Poacher Scenario')
    plt.legend()

    plt.show()

result_plot(data)
```
